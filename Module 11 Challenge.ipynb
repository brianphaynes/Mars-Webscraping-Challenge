```python
 # Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
# Initialize the browser
browser = Browser('chrome', executable_path=r"C:\Users\Brian Haynes\Downloads\chromedriver_win32\chromedriver.exe")

# Visit the Mars Facts site
url = 'https://static.bc-edx.com/data/web/mars_news/index.html'
browser.visit(url)

# Get the HTML code
html = browser.html
```

```python
# Create a BeautifulSoup object from the scraped HTML
mars_soup = soup(html, 'html.parser')
```

```python
# Find all text
all_text = mars_soup.find_all('div', class_ = "list_text")
all_text
```

```python
mars_list = []
```

```python
for article in all_text:
# Extract the title and preview text from the elements
    title = mars_soup.find("div", class_='content_title').text
    preview = mars_soup.find('div', class_="article_teaser_body").text
    # Store each title and preview pair in a dictionary
    text_dict = {}
    text_dict['title'] = title
    text_dict['preview'] = preview
# Add the dictionary to the list
    mars_list.append(text_dict)
```

```python
mars_list
```

```python
import pandas as pd
mars_df = pd.DataFrame(mars_list)
mars_df
```
